# Looking at top 5 topics and which % of all docs they cover

```python
import pandas as pd
import numpy as np
```

```python
reg = ["HI", "ID", "IN", "IP", "LY", "NA", "OP", "SP"]
gen = ["CookbooksFood&Wine", "Engineering&Transportation", "Literature&Fiction", "Medicine&HealthSciences", "Politics&SocialSciences", "Science&Math"]
```

```python
def find_coverage(supports, limit):
    best_topics = supports[1:limit+1]
    all_docs = np.sum(supports)
    return all_docs, np.sum(best_topics)/all_docs
```

```python
from itertools import product

topic_limit=10
support_limit=500
collect = [] #{"label":[],"support":[], "topic_names":[], "coverage":[]}

mean_coverage=0
i = 0
for r, g in product(reg, gen):
    try:
        df = pd.read_csv("topics/"+r+"-"+g+".csv")
    except:
        continue
    support, class_coverage = find_coverage(np.array(df["Count"]),topic_limit)
    if support < support_limit:
        continue
    if len(np.array(df["Name"])[1:1+limit].tolist()) < 5:
        continue
    d = {}
    d["register"] = r
    d["genre"] = g
    d["support"] = support
    d["topic_names"] = np.array(df["Name"])[1:1+limit].tolist()
    d["coverage"] = class_coverage
    collect.append(d)
    mean_coverage += class_coverage
    i+=1
print(mean_coverage/i)
```

```python
df = pd.DataFrame.from_dict(collect)
```
