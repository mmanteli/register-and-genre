## Reading all results cross-labelwise

```python
import re
import json

pattern = re.compile(r'"\w+"')

data = []

with open("1topics_10words_large_data.txt", "r") as f1, open("2topics_10words_large_data.txt", "r") as f2, open("3topics_10words_large_data.txt", "r") as f3:
    t1 = " ".join(f1.readlines()).split("---------------------------")
    t2 = " ".join(f2.readlines()).split("---------------------------")
    t3 = " ".join(f3.readlines()).split("---------------------------")
    for e,(i,j,k) in enumerate(zip(t1, t2, t3)):
        d = {}
        if e == 0:
            continue
        label = i.split("\n")[1][9:]
        support = i.split("\n")[2]
        if "MT" in label:
            continue
        if "Not enough data" in support:
            continue
        #print(f'Label {label[9:-1]}')
        d["register"] = "".join(label.split(",")[0])
        d["genre"] = "".join(label.split(",")[1])
        #print(f'{support}')
        d["support"] = support[12:]
        #print("1 TOPIC:")
        topics1 = pattern.findall("\n".join(i.split("\n")[3:]))
        d["topics1"] = [i[1:-1] for i in topics1]
        #print([i[1:-1] for i in topics1])
        #print("2 TOPICS:")
        topics2 = pattern.findall("\n".join(j.split("\n")[3:]))
        d["topics2"] = []
        for i in [0,10]:
            #print([i[1:-1] for i in topics2[i:i+10]])
            d["topics2"].append([i[1:-1] for i in topics2[i:i+10]])
        #print("3 TOPICS:")
        d["topics3"] = []
        topics3 = pattern.findall("\n".join(k.split("\n")[3:]))
        for i in [0,10,20]:
            #print([i[1:-1] for i in topics3[i:i+10]])
            d["topics3"].append([i[1:-1] for i in topics3[i:i+10]])
        #print("-----------------------")
        data.append(d)

print(json.dumps(data, indent=4))
    
```
